```python
import numpy as np

# Dùng quy tắc 3-sigma kiểm tra Outliers
def check_outliers(column):
    mean = np.mean(column)
    std = np.std(column)
    upper_limit = mean + 3 * std
    lower_limit = mean - 3 * std
    z_scores = [(x - mean) / std for x in column]
    outliers = [x for x in column if x > upper_limit or x < lower_limit]
    return outliers, z_scores

exclude = ['Youtuber', 'Country', 'Channel Type']
count_outliers = 0

for col in df.columns:
    if col not in exclude:
        outliers, z_scores = check_outliers(df[col])
        if len(outliers) > 0:
            count_outliers += len(outliers)
            print(f'Có {len(outliers)} outliers trong cột {col}')
        else:
            print(f'Không có outliers trong cột {col}')

print(f'Có tổng cộng {count_outliers} outliers trong bộ dữ liệu')
```

```python
shape = df.shape
print("Số hàng:", shape[0])
print("Số cột:", shape[1])
```

```python
#Vẽ outliers:
def draw_outliers(column, ax):
    ax.boxplot(column, vert=False)
    # Check outlier và đánh dấu
    outliers, _ = check_outliers(column)
    if len(outliers) > 0:
        ax.scatter(outliers, [1] * len(outliers), c='r', marker='o', label='Outliers')
    mean = np.mean(column)
    std = np.std(column)
    upper_limit = mean + 3 * std
    lower_limit = mean - 3 * std

    # Vẽ đường với các màu cụ thể
    ax.axvline(mean, color='y', linestyle='--')
    ax.axvline(lower_limit, color='g', linestyle='--')
    ax.axvline(upper_limit, color='g', linestyle='--')


fig, axs = plt.subplots(4, 4, figsize=(28, 12))
fig.suptitle('Biểu đồ kiểm tra outliers', fontsize=18)

selected_col = []
for col in df.columns:
    if col not in exclude:
        selected_col.append(col)

for i, col in enumerate(selected_col):
      draw_outliers(df[col], axs[i//4, i%4])
      axs[i//4, i%4].set_title(col)

# Tạo chú thích
red_dot = mlines.Line2D([], [], color='red', marker='o', linestyle='--', label='Outliers')
green_line = mlines.Line2D([], [], color='green', linestyle='--', label='Limit')
yellow_line = mlines.Line2D([], [], color='yellow', linestyle='--', label='Mean')
fig.legend(handles=[red_dot, green_line, yellow_line], loc='upper left', bbox_to_anchor=(0, 1), borderaxespad=0.)


plt.tight_layout()
plt.show()

```

Nhận xét: Mục đích của bài nghiên cứu là tìm ra các insights hữu ích từ các kênh Youtube nổi tiếng trên thế giới, có thể xem xét không loại bỏ outliers ở các cột trên, vì trong thực tế, vẫn tồn tại các trường hợp có các kênh Youtube với lượt xem, lượt đăng ký, thu nhập,... cao hơn hẳn các kênh còn lại.

## Xử lý đặc trưng


Từ năm thành lập kênh (created year), ta có thể tính tuổi theo năm của kênh.

```python
def create_feature(df: pd.DataFrame) -> pd.DataFrame:
    # Copy dataframe sang X
    X = df.copy()

    # Ghép ngày tháng năm vào 1 cột Created Channel
    X['Created Channel'] = X[
        ['Created Year', 'Created Month','Created Date']].astype(str).apply(lambda x: '/'.join(x),axis=1)
    X['Created Channel'] = pd.to_datetime(X['Created Channel'], format="%Y/%m/%d")

    # Tính tuổi (theo năm) bằng cách lấy ngày hôm nay trừ cho Created Channel
    now = datetime.now() # hôm nay
    X['Account Lifetime'] = (now - X['Created Channel']).dt.days # Tính ngày
    X['Account Lifetime'] = (X['Account Lifetime'] / 365).astype(int) # Lấy ngày chia 365 thành năm

    return X
```

##Phân Tích Tương Quan (Correlations)

```python
# Tạo ma trận tương quan Pearson giữa các biến
correlation_matrix = df.corr()
#Chỉnh size
plt.figure(figsize=(25, 17))
# Định nghĩa bảng màu
cmap = sns.color_palette("YlGn", as_cmap=True)
# Vẽ heatmap
sns.heatmap(correlation_matrix, annot=True,cmap=cmap, square=True)
plt.title('Biểu đồ thể hiện tương quan giữa các biến định lượng')
plt.savefig('heatmap.pdf', format='pdf', bbox_inches='tight')
plt.show()
```

```python
correlation_summary = {
    'Tương quan mạnh (+)': set(),
    'Tương quan mạnh (-)': set(),
}

for i, column in enumerate(correlation_matrix.columns):
    for j, row in enumerate(correlation_matrix.index):
        if i < j:  # Tránh việc xem xét cả bên trên và bên dưới đường chéo
            value = correlation_matrix.iloc[j, i]
            if value >= 0.7:
                correlation_summary['Tương quan mạnh (+)'].add((row, column))
            elif value < -0.7:
                correlation_summary['Tương quan mạnh (-)'].add((row, column))

# In kết luận
for key, value in correlation_summary.items():
    print(f'\n')
    for correlation in value:
        print(f"Có sự {key} giữa biến {correlation[0]} và {correlation[1]}")
```

Điều này có thể phản ánh cả chất lượng nội dung, chiến lược tiếp thị, và các yếu tố khác như sự thu hút của các kênh có nhiều người đăng ký và lượt xem cao.

Các giá trị outliers phản ánh những kênh có hoạt động sản xuất nội dung cực kỳ năng suất, có thể góp phần vào sự tăng trưởng nhanh chóng về số lượng người đăng ký và số lượt xem.

```python
# Tính toán hệ số tương quan Pearson giữa các cặp biến quan trọng
correlation_uploads_lowest_earnings      = df['Uploads'].corr(df['Lowest Yearly Earnings'])
correlation_uploads_highest_earnings     = df['Uploads'].corr(df['Highest Yearly Earnings'])
correlation_subscribers_lowest_earnings  = df['Subscribers'].corr(df['Lowest Yearly Earnings'])
correlation_subscribers_highest_earnings = df['Subscribers'].corr(df['Highest Yearly Earnings'])
correlation_views_lowest_earnings        = df['Video Views'].corr(df['Lowest Yearly Earnings'])
correlation_views_highest_earnings       = df['Video Views'].corr(df['Highest Yearly Earnings'])
correlation_subscribers_views            = df['Subscribers'].corr(df['Video Views'])

# Lưu hệ số tương quan vào một từ điển
correlation_coefficients = {
    'Số Video Đã Tải Lên vs Thu Nhập Ước Tính Thấp Nhất': correlation_uploads_lowest_earnings,
    'Số Video Đã Tải Lên vs Thu Nhập Ước Tính Cao Nhất': correlation_uploads_highest_earnings,
    'Số Người Đăng Ký vs Thu Nhập Ước Tính Thấp Nhất': correlation_subscribers_lowest_earnings,
    'Số Người Đăng Ký vs Thu Nhập Ước Tính Cao Nhất': correlation_subscribers_highest_earnings,
    'Lượt Xem Video vs Thu Nhập Ước Tính Thấp Nhất': correlation_views_lowest_earnings,
    'Lượt Xem Video vs Thu Nhập Ước Tính Cao Nhất': correlation_views_highest_earnings,
    'Số Người Đăng Ký vs Lượt Xem Video': correlation_subscribers_views,
}

# Trực quan hóa các mối quan hệ bằng biểu đồ phân tán
# Thiết lập kích thước cho các biểu đồ
plt.figure(figsize=(20, 18))

# Biểu đồ phân tán giữa Số Video Đã Tải Lên và Thu Nhập Ước Tính Thấp Nhất
plt.subplot(4, 2, 1)
sns.scatterplot(x='Uploads', y='Lowest Yearly Earnings', data=df)
plt.title('Số Video Đã Tải Lên vs Thu Nhập Ước Tính Thấp Nhất')
plt.xscale('log')  # Sử dụng thang logarit do phạm vi giá trị rộng
plt.yscale('log')
plt.xlabel('Số Video Đã Tải Lên (thang logarit)')
plt.ylabel('Thu Nhập Ước Tính Thấp Nhất (thang logarit)')

# Biểu đồ phân tán giữa Số Video Đã Tải Lên và Thu Nhập Ước Tính Cao Nhất
plt.subplot(4, 2, 2)
sns.scatterplot(x='Uploads', y='Highest Yearly Earnings', data=df)
plt.title('Số Video Đã Tải Lên vs Thu Nhập Ước Tính Cao Nhất')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Số Video Đã Tải Lên (thang logarit)')
plt.ylabel('Thu Nhập Ước Tính Cao Nhất (thang logarit)')

# Biểu đồ phân tán giữa Số Người Đăng Ký và Thu Nhập Ước Tính Thấp Nhất
plt.subplot(4, 2, 3)
sns.scatterplot(x='Subscribers', y='Lowest Yearly Earnings', data=df)
plt.title('Số Người Đăng Ký vs Thu Nhập Ước Tính Thấp Nhất')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Số Người Đăng Ký (thang logarit)')
plt.ylabel('Thu Nhập Ước Tính Thấp Nhất (thang logarit)')

# Biểu đồ phân tán giữa Số Người Đăng Ký và Thu Nhập Ước Tính Cao Nhất
plt.subplot(4, 2, 4)
sns.scatterplot(x='Subscribers', y='Highest Yearly Earnings', data=df)
plt.title('Số Người Đăng Ký vs Thu Nhập Ước Tính Cao Nhất')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Số Người Đăng Ký (thang logarit)')
plt.ylabel('Thu Nhập Ước Tính Cao Nhất (thang logarit)')

# Biểu đồ phân tán giữa Lượt Xem Video và Thu Nhập Ước Tính Thấp Nhất
plt.subplot(4, 2, 5)
sns.scatterplot(x='Video Views', y='Lowest Yearly Earnings', data=df)
plt.title('Lượt Xem Video vs Thu Nhập Ước Tính Thấp Nhất')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Lượt Xem Video (thang logarit)')
plt.ylabel('Thu Nhập Ước Tính Thấp Nhất (thang logarit)')

# Biểu đồ phân tán giữa Lượt Xem Video và Thu Nhập Ước Tính Cao Nhất
plt.subplot(4, 2, 6)
sns.scatterplot(x='Video Views', y='Highest Yearly Earnings', data=df)
plt.title('Lượt Xem Video vs Thu Nhập Ước Tính Cao Nhất')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Lượt Xem Video (thang logarit)')
plt.ylabel('Thu Nhập Ước Tính Cao Nhất (thang logarit)')

# Biểu đồ phân tán giữa Số Người Đăng Ký và Lượt Xem Video
plt.subplot(4, 2, 7)
sns.scatterplot(x='Subscribers', y='Video Views', data=df)
plt.title('Số Người Đăng Ký vs Lượt Xem Video')
plt.xscale('log')  # Sử dụng thang logarit do phạm vi giá trị rộng
plt.yscale('log')
plt.xlabel('Số Người Đăng Ký (thang logarit)')
plt.ylabel('Lượt Xem Video (thang logarit)')

# Điều chỉnh layout để không bị chồng lên nhau
plt.tight_layout(pad=3.0)
plt.show()

# In ra hệ số tương quan để xem xét
for pair, coefficient in correlation_coefficients.items():
    print(f"{pair}: {coefficient:.3f}")
```

```python
#vẽ biểu đồ top 10 youtubers có nhiều lượt đăng ký nhất
def top_10_bar_chart(df, x_axis, y_axis, chart_title, is_ascending):
  # Sắp xếp DataFrame dựa trên cột y_axis
    df.sort_values(y_axis, ascending=is_ascending, inplace=True)
   # Tạo biểu đồ cột
    fig = px.bar(df.head(10),
        x=x_axis,
        y=y_axis,
        title=chart_title,
        color=x_axis,
    ).update_layout(width=1200, height=800, legend=dict(
        x=1,
        y=1
    ))
# chú thích hiển thị giá trị trung bình
    fig.add_hline(
        y=df[y_axis].mean(),
        line_dash='dash',
        line_color='grey',
        annotation_text=f"Mean: {df[y_axis].mean():.2f}",
        annotation_position="top right"
    )
    return fig
fig_top_subscribers = top_10_bar_chart(df, 'Youtuber', 'Subscribers', 'Top 10 Youtubers nhiều lượt đăng ký nhất', False)
fig_top_subscribers.show()

```

```python
#Xóa outliers
def remove_outliers(df, columns):
    df_cleaned = df.copy()
    for col in columns:
        outliers, _ = check_outliers(df_cleaned[col])
        df_cleaned = df_cleaned[~np.isin(df_cleaned[col], outliers)]
    df_cleaned = df_cleaned.dropna()
    return df_cleaned
```

```python
from kmodes.kprototypes import KPrototypes
cost = []
for k in range(2,15):
  kproto = KPrototypes(n_clusters = k, init = 'Cao')
  clusters = kproto.fit_predict(D, categorical = [0])
  #print('Số clusters:', k)
  #print('     cost (khoảng cách các điểm dữ liệu đến trọng tâm cụm):', kproto.cost_)
  cost.append(kproto.cost_)

plt.xlabel('Số clusters k')
plt.ylabel('Cost')
plt.title('Elbow Method (Tìm số cụm tối ưu)')
plt.plot(cost)
```
